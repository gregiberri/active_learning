```python
# show the class sizes
from data.data_parts import get_datapart
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import csv
import os
import pickle

# # load the data and get the class names and sizes
experiment_names = ['original_label', 'least_uncertainty', 'least_uncertainty_margin', 'max_entropy', 
                    'random', 'full_training_set']
f1_scores = [0.647, 0.778, 0.781, 0.771, 0.767,0.885]


# # plot
fig = plt.figure(figsize=[10, 5])
plt.bar(experiment_names, f1_scores)
plt.xticks(rotation=90)
plt.title('F1 scores')
plt.xlabel('experiment')
plt.ylabel('F1 score')
for i in range(len(f1_scores)):
    plt.annotate(str(f1_scores[i]), 
                 xy=(experiment_names[i],f1_scores[i]), 
                 ha='center', va='bottom')
plt.show()
```
